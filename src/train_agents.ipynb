```python
from training import train_agent
from testing import test_agents
import matplotlib.pyplot as plt
```

```python
LAYOUTS = [
    "cramped_room",
    "asymmetric_advantages",
    "coordination_ring",
    "forced_coordination",
    "counter_circuit_o_1order",
]
```

```python
training_stats_single = []

for layout in LAYOUTS:

    training_stats_single.append(
        train_agent(
            layout,
            episodes=5000,
            model_name=f"{layout}_selfplay"
        )
    )
    training_stats_single[-1]['layout_name'] = layout
```

```python
# Plot the training results
plt.figure(figsize=(12, 6))
for stats in training_stats_single:
    plt.plot(stats["mean_rewards"], label=f"{stats['layout_name']}")
    plt.xlabel("Batch")
    plt.ylabel("Reward")
    plt.title("Average Rewards per Batch")
    plt.legend()

plt.axhline(y=50, color='black', linestyle='--')
plt.show()
```

```python
for layout in LAYOUTS:
    test_agents(
        agents=[f"{layout}_selfplay"],
        layouts=layout,
        verbose=True
    )
```

```python
LAYOUTS = [
    "cramped_room",
    "asymmetric_advantages",
    "coordination_ring",
    "counter_circuit_o_1order",
]


training_stats_rand = []

for layout in LAYOUTS:

    training_stats_rand.append(
        train_agent(
            layout,
            episodes=10000,
            random_prob=0.75,
            model_name=f"{layout}_random")
    )
    training_stats_rand[-1]["layout_name"] = layout


```

```python
# Plot the training results
plt.figure(figsize=(12, 6))
for stats in training_stats_rand:
    plt.plot(stats["mean_rewards"], label=f"{stats['layout_name']}")
    plt.xlabel("Batch")
    plt.ylabel("Reward")
    plt.title("Average Rewards per Batch")
    plt.legend()

plt.axhline(y=50, color="black", linestyle="--")
plt.show()
```

```python
for layout in LAYOUTS:
    test_agents(
        agents=[f"{layout}_random"],
        layouts=layout,
        verbose=True,
    )
    for partner in ['random', 'idle']:
        test_agents(
            agents=[f"{layout}_random", partner],
            layouts=layout,
            verbose=True,
        )
        test_agents(
            agents=[partner, f"{layout}_random"],
            layouts=layout,
            verbose=True,
        )
```
